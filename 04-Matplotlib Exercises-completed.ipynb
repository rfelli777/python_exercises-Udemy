## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig=plt.figure(figsize=(4.5,3.5))

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,'b',lw=1)
axes.set_xlabel('x',fontsize=7)
axes.set_ylabel('y',fontsize=7)
axes.set_title('title',fontsize=10)

#more formatting
axes.set_yticks(np.arange(0,250,step=50))

axes.tick_params(direction='in',labelsize=7,top=True,right=True)
axes.set_xlim(xmin=0,xmax=100)
axes.set_ylim(ymin=0,ymax=200)
```

** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
#group and visualize to connect concepts

#create figure
fig = plt.figure(figsize=(4.5,3.5))

#1st graph
ax1 = fig.add_axes([0,0,1,1])

#formatting for 1st graph
ax1.set_yticks(np.arange(0,1.2,step=.2))

ax1.tick_params(direction='in',labelsize=7,top=True,right=True)
ax1.set_xlim(xmin=0,xmax=1)
ax1.set_ylim(ymin=0,ymax=1)

#2nd graph
ax2 = fig.add_axes([0.2,0.5,.2,.2])

#formatting for 2nd graph
ax2.set_xticks(np.arange(0,1.2,step=.2))
ax2.set_yticks(np.arange(0,1.2,step=.2))

ax2.tick_params(direction='in',labelsize=7,top=True,right=True)
ax2.set_xlim(xmin=0,xmax=1)
ax2.set_ylim(ymin=0,ymax=1)
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
plt.figure(figsize=(3,3))
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(4.3,3))

for ax in axes:
    ax.tick_params(direction='in',labelsize=7,top=True,right=True)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(4,3))

#1st graph
axes[0].plot(x,y,'b--',linewidth=2)

#formatting
axes[0].tick_params(direction='in',labelsize=7)
axes[0].set_xticks(np.arange(0,120,step=20))

#2nd graph
axes[1].plot(x,z,'r-',linewidth=3)

#formatting
axes[1].tick_params(direction='in',labelsize=7)
axes[1].set_ybound(upper=10000)
axes[1].set_xticks(np.arange(0,120,step=20))
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8.5,2))

#first graph
axes[0].plot(x,y,'b-',linewidth=3)
axes[0].set_xlabel('X',fontsize=7)
axes[0].set_ylabel('y',fontsize=7)

#formatting
axes[0].tick_params(direction='in',labelsize=7) #controls tick direction & label font size
axes[0].set_yticks(np.arange(0,250,step=50)) #controls # of ticks

#2nd graph
axes[1].plot(x,z,'r--',linewidth=2)
axes[1].set_xlabel('X',fontsize=7)
axes[1].set_ylabel('Z',fontsize=7)

#formatting
axes[1].tick_params(direction='in',labelsize=7)            
axes[1].set_yticks(np.arange(0,12000,step=2000))

plt.tight_layout()                                 # makes sure axes are properly spaced
```
