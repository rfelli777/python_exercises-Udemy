```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

# Exercises

** Recreate the plots below using the titanic dataframe. There are very few hints since most of the plots can be done with just one or two lines of code and a hint would basically give away the solution. Keep careful attention to the x and y labels for hints.**

** *Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!* **

```python
sns.jointplot(x='fare',y='age',data=titanic,edgecolor='grey')
```

```python
plt.figure(figsize=(4,4)) #resize grid
plt.xlim([0,600]) #started graph from (0,0)
sns.histplot(data=titanic,x='fare',bins=30,color='red',edgecolor=('grey',0.35),alpha=.35)
```

```python
sns.catplot(kind='box',data=titanic,x='class',y='age',palette='rainbow')
```

```python
sns.swarmplot(data=titanic,x='class',y='age',palette='Set2')
```

```python
sns.histplot(data=titanic,x='sex',hue='sex',palette='viridis',alpha=1) #color = close enough
```

```python
sns.heatmap(titanic.corr(numeric_only=True),cmap='coolwarm',vmin=-1)# #vmin = sets range to value of -1
plt.title('titanic.corr')
```

```python
g= sns.FacetGrid(data=titanic,col="sex")
g= g.map(sns.histplot,'age',alpha=1,bins=10,edgecolor='black',linewidth=.5,color='#5471AB') #syntax changed plt.hist--> (changed to sns.histplot)
```
